# Ejericio 4: Vectorizaci√≥n

```python
# Seed para reproducibilidad
import matplotlib.pyplot as plt
from time import time
np.random.seed(42)
# Creamos un arreglo que en su mayoria contiene Nans con algunos "puntos de inflexion" (local min/max).
precios = np.full(100, fill_value=np.nan)
precios[[0, 25, 60, -1]] = [80., 30., 75., 50.]

# De manera lineal interpolamos los valores faltantes y agregamos ruido.
x = np.arange(len(precios))
es_valido = ~np.isnan(precios)
precios = np.interp(x=x, xp=x[es_valido], fp=precios[es_valido])
precios += np.random.randn(len(precios)) * 2
precios
plt.plot(precios)
plt.show()
print(np.amax(precios),"Valor maximo del array")
print(np.amin(precios),"Valor maximo del array")
```

```python
# Usando ciclo For

start_time = time()
def calcular(prices):
    costo = 0
    valor_maximo = 0
    valor_minimo = prices[0]
    if (len(prices) == 0): # devuelve 0 en caso de que no se encuentre ganancia.
        return 0
    for i in range(len(prices)):
        valor_minimo = min(valor_minimo, prices[i])
        costo = prices[i] - valor_minimo
        valor_maximo = max(valor_maximo, costo)
    return valor_maximo
print(f'La ganancia maxima es {calcular(precios)}')
tiempo=time() - start_time
print(f'El tiempo de ejecucion es {tiempo} segundos')
    
```

```python
# Usando Vectorizacion
start_time2 = time()
Ganancia=np.amax(precios[np.where(precios == np.amin(precios))[0][0]:])-np.amin(precios)
print(f'La ganancia maxima es {Ganancia}')
tiempo=time() - start_time2
print(f'El tiempo de ejecucion es {tiempo} segundos')
```

En ambos casos se obtiene la misma respuesta. pero no se puede medir exactamente cual es mas rapido dado que el conjunto de datos es muy peque;o, por la experiencia obtenida en clase usando vectorizacion el rendimiento es mucho mayor que usando iteraciones normales (ciclos for). el segundo codigo arroja un valor correcto por la naturaleza de los datos utilizados. si se utilizara el ejemplo : precios = (20, 18, 14, 17, 20, 21, 13) en el primer codigo la respuesta seria 7 pero en el segundo codigo seria 0 ya que encuentra el valor como 13 y despyes de 13 no hay mas valores maximo
